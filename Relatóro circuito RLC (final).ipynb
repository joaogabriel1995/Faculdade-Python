<b>Modelagem</b>

<p>De acordo com as leis de Kirchoff, a soma de todas as correntes que chegam a um nó do circuito deve ser igual a soma de todas as correntes que deixam esse mesmo nó.</p>
<p>Conforme a lei de Kirchoff das correntes poderemos escrever a seguinte equação para o nó.</p>

$$\displaystyle I_{c} + I_{l} + I_{r} = 0 $$ 
<p>Substituiremos na equação acimas as formulas de corrente para cada componente.</p>
<p>Fórmula para corrente no resistor:</p>
$$\displaystyle I_{r}=\frac{v{\left(t \right)}}{R}$$
<p>Fórmula para corrente no indutor:</p>
$$\displaystyle I_{l}=\frac{v{\left(t \right)}}{R}$$
<p>Fórmula para corrente no capacitor:</p>
$$\displaystyle I_{c}=C\frac{d}{d t} v{\left(t \right)}$$ 
<p>Realizando a substituição encontramos a seguinte equação:</p>
$$\displaystyle C \frac{d}{d t} v{\left(t \right)} + V_{0} + \frac{v{\left(t \right)}}{R} + \frac{\int\limits_{t_{0}}^{t} v{\left(t \right)}\, dt}{L} = 0$$
<p>Podemos perceber que encontramos uma equação integral-derivativa então iremos realizar uma derivada em toda equação, encontrando assim uma equação diferencial de segunda ordem: </p>
$$\displaystyle C \frac{d^{2}}{d t^{2}} v{\left(t \right)} + \frac{\frac{d}{d t} v{\left(t \right)}}{R} + \frac{v{\left(t \right)}}{L} = 0$$
<p>Iremos resolver a equação diferencial, nos apresentando as seguintes
caracteristicas:</p>

$$\displaystyle v{\left(t \right)} = C_{1} e^{\frac{t \left(-1 - \frac{\sqrt{L \left(- 4 C R^{2} + L\right)}}{L}\right)}{2 C R}} + C_{2} e^{\frac{t \left(-1 + \frac{\sqrt{L \left(- 4 C R^{2} + L\right)}}{L}\right)}{2 C R}}$$


$$\displaystyle s_{1} = \frac{-1 - \frac{\sqrt{L \left(- 4 C R^{2} + L\right)}}{L}}{2 C R}$$

$$\displaystyle s_{2} = \frac{-1 + \frac{\sqrt{L \left(- 4 C R^{2} + L\right)}}{L}}{2 C R}$$

Resolução do circuito com valores propostos pelo exercício 1a 

<p>Iremos substituir os valores de R, L e C nas equações s1 e s2 encontrando assim os valores das raízes</p><br>
$$\displaystyle s_1 = -5.40172833186803$$
$$\displaystyle s_2 = −1.54271611257642$$

<p> Podemos então substituir os valores de s1 e s2 na equação da tensão do nosso sistema. </p><br>
$$\displaystyle v{\left(t \right)} = C_{1} e^{- 5.40172833186803 t} + C_{2} e^{- 1.54271611257642 t}$$

<p>Sabemos que no instante de tempo igual a zero segundos a nossa tensão será igual a zero também.</p><br>
$$\displaystyle v{\left(0 \right)} = 0$$
<p>Aplicando essa condição na nossa formula te tensão encontraremos uma equação em função de C1 e C2</p><br>
$$\displaystyle v{\left(0 \right)} = C_{1} e^{- 5,40172833186803 * 0} + C_{2} e^{- 1,54271611257642 * 0}$$<br>
$$\displaystyle  C_{1} + C_{2} = 0$$



<p>Sabemos que $$\displaystyle  I_{r} = 0 $$ </p>
<p>Sabemos que a corrente que sai do indutor irá para o capacitor portanto serão correntes iguais. </p>
$$\displaystyle I_c = C \frac{d}{d t} v{\left(t \right)}$$<br>
$$\displaystyle  \frac{d}{d t} v{\left(t \right)} = 208.33$$
Para encontrarmos a segunda equação precisamos derivar a formula da tensão.
$$\displaystyle \left. \frac{d}{d t} \left(C_{1} e^{- 6.0 t} + C_{2} e^{- 1.0 t}\right) \right|_{\substack{ t=0 }} = 208.33$$
Encontraremos a seguinta equação:<br>
$$\displaystyle - 5.40172833186803 C_{1} - 1.54271611257642 C_{2} = 208.333333333333$$
Resolvendo o sistema linear
$$\displaystyle - 5.40172833186803 C_{1} - 1.54271611257642 C_{2} = 208.333333333333$$
$$\displaystyle  C_{1} = -53.98$$
$$\displaystyle C_{2} = 53.98$$

$$\displaystyle v{\left(t \right)} = - 53.9861813061521 e^{- 5.40172833186803 t} + 53.9861813061521 e^{- 1.54271611257642 t}$$
