<h3>Cosseno e Seno<h3>

```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

```python
#Entendendo o comportamento do seno e cosseno.
def funcao():
    vetort = np.linspace(0,12,1000)
    sen = np.sin(vetort)
    cos = np.cos(vetort)
    return sen, cos
```

```python
fig, ax = plt.subplots()
ax.plot(np.linspace(0,12,1000) ,funcao()[0], label="seno")
ax.plot(np.linspace(0,12,1000) ,funcao()[1], label="cosseno")
ax.grid()
ax.legend();
```

Podemos perceber que o cosseno é adiantado em noventa graus do seno. Sempre que estamos mexendo com fasor é importante usarmos o cosseno. Se tivermos a formula seno precisaremos fazer as devidas transformações. Mais abaixo realizaremos essas mudanças.

<h4>v(t) = 127sen(120t)</h4>
   

1-Primeiro passo é transformar o seno em cosseno. Teremos o seguinte resultado.

v(t) = 127cos(120t - 90)

Na forma fasorial ficará:

127∠-90º

```python
#definindo nossa função e plotando gráfico
def funcao(t):
    radianos = np.deg2rad(120*t + 90)
    func = 127*np.cos(radianos)
    return func
vetort = np.linspace(0, 7, 1000)
fig, ax = plt.subplots()
ax.plot(vetort ,funcao(vetort), label="v(t)")
ax.grid()
ax.legend();
```

```python
#funcao para trocar de polar para cartesiano 
def pol2cart(rho, phi):
    x = rho * np.cos(phi)
    y = rho * np.sin(phi)
    return(x, y)
radianos = np.deg2rad(0-90)
x, y = pol2cart(127, radianos)
print("{}{}j" .format(round(x),round(y)))
print("forma polar",cmath.polar(complex(x,y)))
```
